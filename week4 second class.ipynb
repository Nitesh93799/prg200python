```python
class Student:
    '''student class with constructer'''

    def __init__(self, name, grade, student_id):
        """
        constructor methode
        'self' refers to the instance being created
        """
        self.name= name
        self.grade= grade
        self.student_id= student_id

    def display_info(self):
        print(f"Student: {self.name}")
        print(f"ID: {self.student_id}")
        print(f"Grade: {self.grade}")
        print("-"*30)

s1=Student('Nitesh',3,'A')
s1.display_info()
```

```python
class Student:

    def __init__(self, name, marks):
        self.name= name
        self.marks= marks

    def display_info(self):
        print(f"Student: {self.name}")
        print(f"Grade: {self.marks}")
        print("-"*30)

s1=Student('Nitesh',60)
s1.display_info()
s2=Student('Ashim',80)
s2.display_info()
s3=Student('Dibya',80)
s3.display_info()
s4=Student('Rajan',90)
s4.display_info()
s5=Student('Sabin',76)
s5.display_info()
s6=Student('Samir',65)
s6.display_info()

students = [s1, s2, s3, s4, s5, s6]

highest = max(students, key=lambda x: x.marks)
lowest = min(students, key=lambda x: x.marks)
average = sum(s.marks for s in students) / len(students)
print("Highest:", highest.name, highest.marks)
print("Lowest:", lowest.name, lowest.marks)
print("Average:", average)
```
