```python
inventory = {
    "rice": {"price": 120, "stock": 20},
    "milk": {"price": 90, "stock": 10},
    "bread": {"price": 60, "stock": 15},
    "eggs": {"price": 15, "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

def process_order(inventory, cart):
    total_bill = 0

    print("---- Bill ----")

    for item, qty in cart.items():
        if item in inventory and inventory[item]["stock"] >= qty:
            cost = inventory[item]["price"] * qty
            total_bill += cost
            inventory[item]["stock"] -= qty

            print(f"{item} x{qty} = NPR {cost}")
        else:
            print(f"Sorry, not enough stock for {item}")

    print(f"\nGrand Total: NPR {total_bill}")
    print("--------------")

    print("\nUpdated stock:")
    for item in inventory:
        print(f"{item} = {inventory[item]['stock']}")

process_order(inventory, cart)
```

```python
sensors = [
    ("Chatara", 2.8),
    ("Tribeni Ghat", 5.4),
    ("Koshi Barrage", 4.1),
    ("Sunsari Bridge", 1.9),
    ("Saptakoshi Camp", 6.0),
]

def check_water_level(location, level_metres):
    if level_metres < 3:
        return "Safe"
    elif level_metres <= 5:
        return "Warning — Alert nearby villages"
    else:
        return "DANGER — Evacuate immediately!"

for location, level in sensors:
    print(f"{location} ({level} m): {check_water_level(location, level)}")
```

```python
bs_months = [
    "Baisakh", "Jestha", "Ashadh", "Shrawan",
    "Bhadra", "Ashwin", "Kartik", "Mangsir",
    "Poush", "Magh", "Falgun", "Chaitra"
]

customers = [
    {"name": "Amesh Thapa", "date": "1985-06-20", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Nita Karki", "date": "2055-09-23", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Rakash Rai", "date": "1998-11-18", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjun Gurung", "date": "2040-01-27", "cal": "BS", "need": "AD", "style": "full"}
]

def convert_date(date_str, from_cal, to_cal):
    year, month, day = map(int, date_str.split("-"))

    if from_cal == to_cal:
        return f"{year:04d}-{month:02d}-{day:02d}"

    if from_cal == "AD" and to_cal == "BS":
        year += 56
    elif from_cal == "BS" and to_cal == "AD":
        year -= 56

    return f"{year:04d}-{month:02d}-{day:02d}"

def format_date(date_str, cal, style):
    year, month, day = map(int, date_str.split("-"))

    if style == "iso":
        return date_str + " " + cal

    elif style in ["full", "nepali"]:
        month_name = bs_months[month - 1]
        return f"{day}th {month_name}, {year} {cal}"

    return date_str + " " + cal


for customer in customers:
    converted = convert_date(
        customer["date"],
        customer["cal"],
        customer["need"]
    )

    formatted = format_date(
        converted,
        customer["need"],
        customer["style"]
    )

    print(
        f'{customer["name"]} | Original: {customer["date"]} {customer["cal"]} | Converted: {formatted}'
    )
```

```python
def word_frequency(text):
    
    for ch in [".", ",", "!", "?"]:
        text = text.replace(ch, "")

    words = text.lower().split()

    freq = {}

    for word in words:
        freq[word] = freq.get(word, 0) + 1

    sorted_words = sorted(freq.items(), key=lambda x: x[1], reverse=True)

    return sorted_words[:3]


text = """
Nepal is a beautiful country. Nepal has Mount Everest. 
Everest is the highest mountain in the world. Many tourists 
visit Nepal every year to see Everest and other mountains. 
Nepal is known for its mountains and natural beauty. 
"""

top3 = word_frequency(text)

print("Top 3 words:")
for word, count in top3:
    print(f"{word} — {count} times")
```

```python
accounts = {
    "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki", "balance": 8500, "pin": "5678"},
    "A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
}

def atm(account_id, pin, action, amount=0):

    if account_id not in accounts:
        print("Account not found")
        return

    account = accounts[account_id]

    if account["pin"] != pin:
        print("Incorrect PIN")
        return

    if action == "balance":
        print(f"{account['name']} | Balance: NPR {account['balance']}")

    elif action == "deposit":
        account["balance"] += amount
        print(f"Deposit successful. New balance: NPR {account['balance']}")

    elif action == "withdraw":
        if amount > account["balance"]:
            print("Insufficient funds")
        else:
            account["balance"] -= amount
            print(f"Withdrawal successful. New balance: NPR {account['balance']}")

    else:
        print("Invalid action")


atm("A001", "1234", "balance")
atm("A002", "0000", "withdraw", 2000)
atm("A002", "5678", "deposit", 3000)
atm("A003", "9012", "withdraw", 25000)
atm("A004", "1111", "balance")
```
