```python
class BankAccount:
    def __init__(self, name, account_number, balance=0):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited Rs.{amount} into {self.account_number}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrawn Rs.{amount} from {self.account_number}")

    def get_balance(self):
        print(f"Name: {self.name}, Balance: Rs.{self.balance}")
```

```python
# Given account data
accounts = [
    ("Nitesh Shahi", "A001", 8000),
    ("Sunil Sarki", "A002", 0),
    ("Braj Roa", "A003", 15000),
]
```

```python
# Create objects
bank_accounts = {}

for name, acc_no, balance in accounts:
    bank_accounts[acc_no] = BankAccount(name, acc_no, balance)
```

```python
# Operations
bank_accounts["A002"].deposit(3000)
bank_accounts["A003"].withdraw(15000)   # Should fail
bank_accounts["A001"].withdraw(2000)
```

```python
# Final balances
print("\nFinal Balances:")
for account in bank_accounts.values():
    account.get_balance()
```
