```python
member = ["Nitesh", "Ashim", "Arun", "Sumit"]

tasks = {
    "Nitesh": "Papers",
    "Ashim": "field visit",
    "Arun": "Documentation",
    "Sumit": "advertisement"
}

expenses = {
    "Nitesh": 1000,
    "Ashim": 4000,
    "Arun": 500,
    "Sumit": 4500
}
```

```python
def show_members(member_list):
    print("Group Members:")
    for m in member_list:
        print("-", m)
```

```python
def show_tasks(task_dict):
    print("\nTasks Assigned:")
    if not task_dict:
        print("No tasks available")
        return

    for name, task in task_dict.items():
        print("-", name, ":", task)
```

```python
def split_expenses(member_list, expense_dict):
    total = sum(expense_dict.values())
    per_person = total / len(member_list)

    print("\nTotal Expense:", total)
    print("Each Person Should Pay:", per_person)

    print("\nBalance Sheet:")
    for person in member_list:
        balance = expense_dict[person] - per_person

        if balance > 0:
            print(person, "will get", balance)
        elif balance < 0:
            print(person, "should pay", abs(balance))
        else:
            print(person, "is paid")
```

```python
def result():
    show_members(member)
    show_tasks(tasks)
    split_expenses(member, expenses)
```

```python
result()
```

```python
students = {}
```

```python
def add_student():
    name = input("Name of Student: ")
    marks = input("Marks of students: ")
    students[name] = marks
    print("Students added!")
```

```python
def view_students():
    for name, marks in students.items():
        print(name, ":", marks)
```

```python
def output():
    while True:
        print("\n1.Add Student  2.View Student  3.Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            break
        else:
            print("Invalid")

output()
```

```python
accounts = {}

def create():
    name = input("Name: ")
    accounts[name] = int(input("Initial deposit: "))
    print("Account created!")
```

```python
def deposit():
    name = input("Name: ")
    amt = int(input("Amount: "))
    if name in accounts:
        accounts[name] += amt
        print("Deposited")
    else:
        print("No account, create from adbl.com.np")
```

```python
def withdraw():
    name = input("Name: ")
    amt = int(input("Amount: "))
    if name in accounts:
        if accounts[name] - amt >= 500:
            accounts[name] -= amt
            print("Withdrawn")
        else:
            print("Minimum 500 required")
    else:
        print("No account,, Please Create From adbl.com")
```

```python
def balance():
    name = input("Name: ")
    print("Balance:", accounts.get(name, "No account, create from adbl.com.np"))
```

```python
def transfer():
    f = input("From: ")
    t = input("To: ")
    amt = int(input("Amount: "))

    if f in accounts and t in accounts:
        if accounts[f] - amt >= 500:
            accounts[f] -= amt
            accounts[t] += amt
            print("Transferred")
        else:
            print("Not enough balance")
    else:
        print("No account, create from adbl.com.np")
```

```python
def view():
    print("\nAccounts:")
    for k, v in accounts.items():
        print(k, ":", v)
```

```python
def result():
    while True:
        print("\n1.Create 2.Deposit 3.Withdraw 4.Balance 5.Transfer 6.View 7.Exit")
        ch = input("Choice: ")

        if ch == "1": create()
        elif ch == "2": deposit()
        elif ch == "3": withdraw()
        elif ch == "4": balance()
        elif ch == "5": transfer()
        elif ch == "6": view()
        elif ch == "7": break
        else: print("Invalid")

result()
```
