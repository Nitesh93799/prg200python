```python
import random

random.seed(42)

friends = ["Ramraja", "Sunil", "Bikash", "Arjun", "Dipsika"]
total_bill = 3750

def split_bill(friends, total):
    return total / len(friends)

def pick_lucky(friends):
    return random.choice(friends)

def final_summary(friends, total):
    share = split_bill(friends, total)
    lucky_person = pick_lucky(friends)

    print("=== Bill Split Summary ===")
    print(f"Total Bill: NPR {total}")
    print(f"Equal Share Per Person: NPR {share:.2f}\n")

    print("Each Person Pays:")
    for person in friends:
        print(f"{person}: NPR {share:.2f}")

    lucky_total = share + 50

    print("\nLucky Person:", lucky_person)
    print(f"{lucky_person} pays an extra NPR 50 as lucky tax.")
    print(f"Total Paid by {lucky_person}: NPR {lucky_total:.2f}")

final_summary(friends, total_bill)
```
