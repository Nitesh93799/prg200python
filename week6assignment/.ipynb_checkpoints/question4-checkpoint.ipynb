```python
%%writefile discount.py

TAX_RATE = 0.13

def apply_discount(price, percent):
    return price - (price * percent / 100)

def apply_tax(price):
    return price + (price * TAX_RATE)

def final_price(price, discount_pct):
    discounted_price = apply_discount(price, discount_pct)
    return apply_tax(discounted_price)
```

```python
from discount import final_price, TAX_RATE

products = [
    ("Laptop", 95000, 10),
    ("Headphones", 5500, 15),
    ("Phone Case", 750, 5),
    ("USB Cable", 650, 0),
]

print("VAT Rate:", TAX_RATE * 100, "%")
print("-" * 40)

for name, price, discount in products:
    final = final_price(price, discount)

    print("Product:", name)
    print("Original Price: NPR", price)
    print("Final Price: NPR {:.2f}".format(final))
    print("-" * 40)
```
